```python
# Factorials --> Continous Multiplications
#Bigger Problem be broken into smaller problems
# 5 Factorial = 1*2*3*4*5
# 7 Factorial = 1*2*3*4*5*6*7
# 6 Factorial = 1*2*3*4*5*6
# 5 Factorial = 1*2*3*4*5
# 4 factorial = 1*2*3*4
# 3 Factorial = 1*2*3
# 2 Factorial = 1*2
# 1 Factorial = 1
# Num factorial = factorial(n-1)*n
def factorial(num):
  if (num==1):
    return 1
  else:
    return factorial(num-1)*num
```
